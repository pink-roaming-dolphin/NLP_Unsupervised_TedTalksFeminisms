```python
ted = pd.read_csv("ted_clean.csv")
```

```python
script = ted.transcript
```

```python
%store -r data
stop_words = data
```

```python
tf = CountVectorizer(stop_words = stop_words, token_pattern = r'\b[a-zA-Z]{3,}\b', \
                     max_df = 0.25, min_df = 250)
```

```python
doc_term_matrix = tf.fit_transform(script)
```

```python
vocab = tf.get_feature_names()
```

```python
def show_topics(num_words= 10): 
    topic_words = []
    for idx, topic in enumerate(tm.components_):
    # Select the top 15 words in vocab for this topic.
        top_words = [vocab[i] for i in topic.argsort()[:-num_words-1:-1]]
        #print(f"Topic {idx}:\n", ", ".join(top_words), "\n")
        topic_words.append(top_words) 
    return topic_words
```

```python
tm = LatentDirichletAllocation(n_components = 15, random_state = 0, doc_topic_prior = 5, topic_word_prior = 2)
tm.fit(doc_term_matrix)
topic_words = show_topics()
```

## PART 2: Look at individual topics in detail

**1.GENDER / FEMINISMS**

```python
feminisms = ted_top[ted_top.topic == 3]
len(feminisms)
```

```python
feminisms.head(5)
```

**1. Associated other topics**

```python
def def_value():
    return "Not Present"
     
d = defaultdict(def_value)
weights = []
counter = []

for i in range(15): 
    print(f"With feminisms, chances of seeing topic {i}: {topic_dict[i]} is {feminisms[i].mean()}.")
    d[feminisms[i].mean()] = topic_dict[i]
    weights.append(feminisms[i].mean())
    counter.append(topic_dict[i])

del weights[3]
del counter[3]
sorted_weights = sorted(weights, reverse=True)[:4]    

print("\n")
print(f"Most likely other topics are: {d[sorted_weights[0]]}, {d[sorted_weights[1]]}, {d[sorted_weights[2]]}.") 
```

```python
g = sns.barplot(x = counter, y = weights)
g.set_xticklabels(g.get_xticklabels(),rotation = 60)
g.set_title("Topics commonly associated with the topic GENDER/FEMINISMS")
```

**2.Views over the years**

```python
g = sns.barplot(x = feminisms.year, y = feminisms.views, ci=None)
g.set_title("Views of talks with topic GENDER / FEMINISMS over the years")
g.set(xlabel=None, ylabel="Number of Views")
```

**3. Comments over the years**

```python
g = sns.barplot(x = feminisms.year, y = feminisms.comments, ci=None)
g.set_title("Commenting on the topic GENDER / FEMINISMS over the years")
g.set(xlabel=None, ylabel="Number of Comments")
```

**2.FOOD SYSTEMS**

```python
food = ted_top[ted_top.topic == 7]
print(f"Number of food systems talks are: {len(food)}.")
food.head(3)
```

**1. Associated other topics**

```python
d_food = defaultdict(def_value)

weights_food = []
counter_food = []
for i in range(15): 
    print(f"With food systems, chances of seeing topic {i}: {topic_dict[i]} is {food[i].mean()}.")
    d_food[food[i].mean()] = topic_dict[i]
    weights_food.append(food[i].mean())
    counter_food.append(topic_dict[i])

del weights_food[7]
del counter_food[7]
    
sorted_weights_food = sorted(weights_food, reverse=True)[:3]
print("\n")
print(f"Most likely other topics are: {d_food[sorted_weights_food[0]]}, {d_food[sorted_weights_food[1]]}, {d_food[sorted_weights_food[2]]}.") 
```
