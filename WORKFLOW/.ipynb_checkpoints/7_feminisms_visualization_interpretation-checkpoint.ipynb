```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# import sklearn
from sklearn.feature_extraction.text import CountVectorizer 
from sklearn.decomposition import LatentDirichletAllocation
```

**Make Gender/ Feminisms DF**

```python
ted_top = pd.read_csv("ted_topicwise.csv")
```

```python
ted_top.head(3)
```

```python
feminisms = ted_top[ted_top.topic == 3]
feminisms.head(3)
```

# EDA Feminisms

This will have 3 parts: 
- Part 1: Talks added
- Part 2: Viewership 
- Part 3: Commentary

## Part 1: Talks added 

**Number of Talks over years**

```python
sns.set(rc={'figure.figsize':(12,9)})
sns.set_style("white")
```

```python
fem_over_yrs = feminisms.groupby('year').sum()
fem_over_yrs.reset_index(inplace=True)
g = sns.barplot(x = fem_over_yrs.year, y= fem_over_yrs.topic)
g.set_xticklabels(g.get_xticklabels(),rotation = 35)
g.set(xlabel=None, ylabel="Number of talks")
g.set_title("Number of talks with topic GENDER/ FEMINISMS over years", fontsize=20)
plt.savefig("Number of talks with topic FEMINISMS over years.png", bbox_inches='tight', dpi = 250)
```

**Number of Talks as percentage of overall topics over Years**

```python
#recreate topics by year
ted_grouped_by_year_topic = ted_top.groupby(['year', 'topic_title']).count()['topic']
ted_grouped_by_year_topic = ted_grouped_by_year_topic.to_frame().reset_index()
ted_grouped_by_year_topic.rename(columns = {'index':'year', 'topic': 'talks_per_topic'},
          inplace = True)

#figure out total number of talks
sum_talks_list = []
years = []
for i in range(1984, 2021): 
    sum_talks_list.append(ted_grouped_by_year_topic[ted_grouped_by_year_topic.year == i].talks_per_topic.sum())
    years.append(i)
sum_talks = dict(zip(years, sum_talks_list))    

#create the columns for percentage 
ted_grouped_by_year_topic["total_talks_that_year"] = ted_grouped_by_year_topic.year.map(lambda x: sum_talks.get(x))
ted_grouped_by_year_topic['topic_percent_that_year'] = ted_grouped_by_year_topic.talks_per_topic/ ted_grouped_by_year_topic.total_talks_that_year
```

```python
ted_grouped_by_year_topic.head(3)
```

```python
palette = {c:'purple' if c=='GENDER / FEMINISM(S)' else 'lightgrey' for c in ted_grouped_by_year_topic.topic_title.unique()}

#now plot this for feminisms 
g = sns.lineplot(x = "year", y = "topic_percent_that_year", hue = "topic_title", \
             data = ted_grouped_by_year_topic[ted_grouped_by_year_topic.year > 2004], \
                palette=palette)
g.set(xlabel=None, ylabel="Percentage of talks")
g.set_xticks(range(2004, 2021))
g.set_title("Percentage of talks given with that topic over years, after 2004", fontsize=20)
plt.savefig("Percentage of talks given with that topic over years, after 2004.png", bbox_inches='tight', dpi = 250)
```

**Probability percentages**

```python
top_percentages = ted_top.iloc[:, 19:-1]
top_percentages.drop(columns = ["topic"], inplace = True)
top_percentages = top_percentages.groupby(['year']).mean()
top_percentages.reset_index(inplace=True)
top_percentages.head(3)
```

```python
g = sns.lineplot(x = 'year', y = "3", data = top_percentages[top_percentages.year > 2004])
g.set(xlabel=None, ylabel="Topic probability")
g.set_title("Topic probability for topic GENDER / FEMINISM(S) over the years", fontsize=20)
g.set_xticks(range(2004, 2021))
plt.savefig("Topic probability for particular topic over the years.png", bbox_inches='tight', dpi = 250)
```

## Part 2: Viewership

**Viewership of the topic GENDER/ FEMINISMS over years**

```python
feminisms[feminisms.year == 2006]
```

```python
g = sns.barplot(x = feminisms.year, y = feminisms.views, ci=None)
g.set(xlabel=None, ylabel="Number of Views")
g.set_title("Views of talks with topic GENDER/ FEMINISMS over the years", fontsize=20)
plt.savefig("Views of talks with topic Feminisms, Distinct.png", bbox_inches='tight', dpi = 250)
```

**Cumulative viewership of topic GENDER/ FEMINISMS over the years**

```python
fem_years = feminisms.groupby(["year"]).sum()
fem_years.reset_index(inplace = True)
fem_years
```

```python
g= sns.barplot(data = fem_years, x = "year", y = "views")
g.set_title("Total viewership of topic GENDER/ FEMINISMS over the years", fontsize=20)
g.set(xlabel=None, ylabel="Number of Views")
plt.savefig("Total viewership of topic Feminisms over the years", bbox_inches='tight', dpi = 250)
```

**Engagement percentage overall by viewership**

```python
sum_views_list = []
years = []
for i in range(1984, 2021): 
    sum_views_list.append(ted_top[ted_top.year == i].views.sum())
    years.append(i)

sum_views = dict(zip(years, sum_views_list))    
    
print(sum_views)
```

```python
sum_fem_views_list = []
years = []
for i in range(1984, 2021): 
    sum_fem_views_list.append(feminisms[feminisms.year == i].views.sum())
    years.append(i)

sum_fem_views = dict(zip(years, sum_fem_views_list))    
    
print(sum_fem_views)
```

```python
feminisms['total_views_for_year'] = feminisms.year.map(lambda x: sum_views.get(x))
#feminisms['comment_perc_that_year'] = feminisms.comments / feminisms['total_comments_for_year']
```

```python
feminisms['topic_views_for_year'] = feminisms.year.map(lambda x: sum_fem_views.get(x))
```

```python
feminisms['percentage_views'] = feminisms['topic_views_for_year'] / feminisms['total_views_for_year']
```

```python
g = sns.lineplot(data = feminisms[feminisms.year > 2004], x = 'year', y = 'percentage_views')
g.set(xlabel=None, ylabel="Percentage")
g.set_title("Viewership in this topic as a percentage of the overall viewership", fontsize=20)
g.set_xticks(range(2004, 2021))
plt.savefig("Viewership in this topic as a percentage of the overall viewership.png", bbox_inches='tight', dpi = 250)
```

**Engagement percentage overall by commenting**

```python
sum_comments_list = []
years = []
for i in range(1984, 2021): 
    sum_comments_list.append(ted_top[ted_top.year == i].comments.sum())
    years.append(i)

sum_comments = dict(zip(years, sum_comments_list))    
    
print(sum_comments)
```

```python
sum_fem_comments_list = []
years = []
for i in range(1984, 2021): 
    sum_fem_comments_list.append(feminisms[feminisms.year == i].comments.sum())
    years.append(i)

sum_fem_comments = dict(zip(years, sum_fem_comments_list))    
    
print(sum_fem_comments)
```

```python
feminisms['total_comments_for_year'] = feminisms.year.map(lambda x: sum_comments.get(x))
#feminisms['comment_perc_that_year'] = feminisms.comments / feminisms['total_comments_for_year']
```

```python
feminisms['topic_comments_for_year'] = feminisms.year.map(lambda x: sum_fem_comments.get(x))
```

```python
feminisms['percentage_comments'] = feminisms['topic_comments_for_year'] / feminisms['total_comments_for_year']
```

```python
g = sns.lineplot(data = feminisms[feminisms.year >2004], x = 'year', y = 'percentage_comments')
g.set(xlabel=None, ylabel="Percentage")
g.set_title("Comments in GENDER / FEMINISM(S) as a percentage of the overall comments", fontsize=20)
g.set_xticks(range(2004, 2021))
plt.savefig("Comments in FEMINISMS as a percentage of the overall comments.png", bbox_inches='tight', dpi = 250)
```
