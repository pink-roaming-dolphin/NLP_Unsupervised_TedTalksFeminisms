## Now onto Tuning the LDA 

```python
def show_topics(num_words= 10): 
    topic_words = []
    for idx, topic in enumerate(tm.components_):
    # Select the top 15 words in vocab for this topic.
        top_words = [vocab[i] for i in topic.argsort()[:-num_words-1:-1]]
        print(f"Topic {idx}:\n", ", ".join(top_words), "\n")
        topic_words.append(top_words) 
    return topic_words
```

**2. learning_method**: Let's also switch to 'online' so that we are a bit faster.

```python
tm = LatentDirichletAllocation(n_components = 15, random_state = 0, learning_method = 'online')
tm.fit(doc_term_matrix)
show_topics()
```

This actually seems to worsen our topics, so let's go back to the default learning_method.

**3. theta & beta**: Let's also see if using prior topic words help our case. 

```python
tm = LatentDirichletAllocation(n_components = 15, random_state = 0, doc_topic_prior = 4)
tm.fit(doc_term_matrix)
show_topics()
```

I also would say we improved slightly. So let's up that doc_topic_prior. 

```python
tm = LatentDirichletAllocation(n_components = 15, random_state = 0, doc_topic_prior = 10)
tm.fit(doc_term_matrix)
show_topics()
```

Too dispersed. Somewhere in between is the answer. A bit experimentation lands us at doc_topic_prior: 5

```python
tm = LatentDirichletAllocation(n_components = 15, random_state = 0, doc_topic_prior = 5)
tm.fit(doc_term_matrix)
show_topics()
```

Now, let's look at topic_word_prior: Experimenting with topic_word_prior = 1, 5, 10, 100; we realize that there are more action words with when we increase the topic_word_prior (beta), we also get more specific words but also looks some of the diversity. It seems that 2 is a sweet spot. 

```python
tm = LatentDirichletAllocation(n_components = 15, random_state = 0, doc_topic_prior = 5, topic_word_prior = 2)
tm.fit(doc_term_matrix)
show_topics()
```

**4. max_iter**: Let's also see if this helps us. \
Experimenting with 5, 10, 20, 50 shows us that actually the default of 10 is quite the best one.  

```python
tm = LatentDirichletAllocation(n_components = 15, random_state = 0, \
                               doc_topic_prior = 5, topic_word_prior = 2, max_iter = 15)
tm.fit(doc_term_matrix)
show_topics()
```

OK, one last time so we can save our params & visualize afterwards.

```python
tm = LatentDirichletAllocation(n_components = 15, random_state = 0, \
                               doc_topic_prior = 5, topic_word_prior = 2)
tm.fit(doc_term_matrix)
topic_words = show_topics()
```
